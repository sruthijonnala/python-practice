```python
# line chart
x = [10,20,30,40,50,60,70]
y= [0,2,3,5,6,8,9]

plt.plot(x,y, marker = 'o', color = 'Green', markerfacecolor = 'red', mec = 'black')
for i, v in zip(x,y):
    plt.text(i, v+0.2, str(v))
plt.xlabel('temperature')
plt.ylabel('units')
plt.grid(True,linewidth = 5)
plt.show()
```

```python
# grouped Bar chart
import numpy as np
x = ['choclate', 'icecream', 'Pizza', 'Samosa']
case1 = [60,45,25,60]
case2 = [55,30,10,60]
case3 = [23,65,20,10]
width = 0.2
xp = np.arange(len(x))
plt.bar(xp - width,case1,width, color = 'Green', label = 'case1')
plt.bar(xp,case2, width, color = 'Red', label = 'case2')
plt.bar(xp + width, case3, width, color = 'm',label = 'case3')
plt.xlabel('food')
plt.ylabel('sales%')
plt.xticks(xp,x)
plt.legend()   
plt.show()
```

```python
# bar chart
sub = ['maths','science','social']
marks = [23,25,24]
clr = ['blue','green','pink']
plt.bar(sub,marks, width = 0.4, color = clr)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.show()
```

```python
sub = ['maths','science','social']
raju = [10,20,30]
john = [19,22,24]
smith = [15,25,24]
width = 0.2
axis = np.arange(len(sub))

plt.bar(axis - width, raju, width = 0.2, color = 'pink', label = 'raju')
plt.bar(axis,john, width = 0.2, color = 'green', label = 'john')
plt.bar(axis + width,smith, width = 0.2, color = 'orange', label = 'smith')
plt.xlabel('subjects')
plt.ylabel('marks')
plt.legend()
plt.xticks(axis,sub)
plt.show()
```
