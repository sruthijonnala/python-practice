```python
import numpy as np
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

data = pd.DataFrame(exam_data,index = labels)
data
```

```python
#Write a Pandas program to select the rows where the score is missing, i.e. is NaN.
data[data['score'].isnull()]
```

```python
# select the rows the score is between 15 and 20
score_between = data[(data['score'] >= 15) & (data['score'] <= 20)]
score_between
```

```python
#select the rows where number of attempts in the examination is less than 2 and score greater than 15.
attempts = data[(data['attempts'] < 2) & (data['score'] >15)]
attempts
```

```python
# calculate the sum of the examination attempts by the students.
total = data['attempts'].sum()
print('sum of attempts:', total)
```
