```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperatures = [22, 25, 27, 26, 28, 30, 29]
plt.plot(days,temperatures, marker = 'o', linestyle = '--', color = 'r', linewidth = 2)
plt.xlabel('days')
plt.ylabel('temp')
plt.title('days vs temp')
plt.grid( linestyle = '--', color = 'blue', linewidth = 0.5, alpha = 0.5)
plt.xticks(days, rotation = 90)
plt.legend(["Temperature (°C)"])
clr = plt.gca()
clr.set_facecolor("black")
max_temp = max(temperatures)
min_temp = min(temperatures)
max_index = temperatures.index(max_temp)
min_index = temperatures.index(min_temp)
plt.scatter(days[max_index],max_temp, marker = '*', color = 'white')
plt.scatter(days[min_index],min_temp, marker = 's', color = 'pink')
plt.show()
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperatures = [22, 25, 27, 26, 28, 30, 29]

# Create figure
plt.figure(figsize=(10, 6))

# Plot line with color change based on temperature threshold (27°C)
for i in range(len(days) - 1):
    if temperatures[i] > 27 or temperatures[i+1] > 27:
        color = 'blue'  # Above 27°C
    else:
        color = 'red'   # 27°C or below
    
    plt.plot(days[i:i+2], temperatures[i:i+2], linestyle="--", marker="o", color=color, linewidth=2)

# Add data labels above each point
for i, temp in enumerate(temperatures):
    plt.text(days[i], temp + 0.5, str(temp), ha='center', fontsize=10, fontweight='bold')

# Highlight max and min temperatures
max_temp = max(temperatures)
min_temp = min(temperatures)
max_index = temperatures.index(max_temp)
min_index = temperatures.index(min_temp)

plt.scatter(days[max_index], max_temp, color="blue", marker="*", s=150, label="Max Temp (30°C)")
plt.scatter(days[min_index], min_temp, color="green", marker="s", s=100, label="Min Temp (22°C)")

# Background shading for temperature ranges
plt.axhspan(0, 27, facecolor='lightcoral', alpha=0.3)   # Light Red for ≤ 27°C
plt.axhspan(27, max_temp + 2, facecolor='lightblue', alpha=0.3)  # Light Blue for > 27°C

# Labels and Title
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)
plt.title("Temperature Variation Over a Week", fontsize=14, fontweight="bold")

# Rotate x-axis labels
plt.xticks(rotation=30)

# Grid lines
plt.grid(True, linestyle="--", alpha=0.5)

# Legend
plt.legend()

# Show the chart
plt.show()
```

```python
category = ['Electronics','Clothing', 'Groceries', 'Furniture', 'Sports']
sales = [120,85,150,90,75]
clr = ['red','green','black','violet','pink']
plt.bar(category, sales, width = 0.6, color = clr, edgecolor = 'white')
plt.xlabel('Product Category', fontsize = 14)
plt.ylabel('Sales (in $1000s)',fontsize = 14)
plt.title('Retail Store Sales by Category', fontsize = 16, fontweight = 'bold')
plt.grid(axis = 'y',linestyle = '--', alpha = 0.7)
plt.xticks(rotation = 30, fontsize = 13)
plt.gca().set_facecolor('grey')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend(["Sales Data"], loc='upper right', fontsize=12, frameon=True)
plt.show()
```

```python
categories = ["Electronics", "Clothing", "Groceries", "Furniture", "Sports"]
sales = [120, 85, 150, 90, 75]
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700']  # Custom colors

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bar chart with customization
bars = ax.bar(categories, sales, color=colors, edgecolor='black', width=0.6)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}', 
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Title and labels
ax.set_title("Retail Store Sales by Category", fontsize=16, fontweight='bold')
ax.set_xlabel("Product Category", fontsize=14)
ax.set_ylabel("Sales (in $1000s)", fontsize=14)

# Rotate x-axis labels
ax.set_xticklabels(categories, rotation=30, fontsize=12)

# Add grid lines
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Customize spines (remove top & right)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set background color
ax.set_facecolor('#f5f5f5')

# Add legend
ax.legend(["Sales Data"], loc='upper right', fontsize=12, frameon=True)

# Show the plot
plt.show()
```

```python
categories = ["Electronics", "Clothing", "Groceries", "Furniture", "Sports"]
sales = [120, 85, 150, 90, 75]
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700']  # Custom colors

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 7))

# Plot bar chart with customization
bars = ax.bar(categories, sales, color=colors, edgecolor='black', width=0.6, alpha=0.85, linewidth=1.5)

# Add shadow effect
for bar in bars:
    bar.set_linewidth(1.5)
    bar.set_edgecolor('black')
    bar.set_zorder(3)  # Ensures bars are above the grid

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 2, f'${yval}K', 
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Title and labels
ax.set_title("Retail Store Sales by Category", fontsize=18, fontweight='bold', color='#333')
ax.set_xlabel("Product Category", fontsize=14, fontweight='bold', labelpad=10, color='#444')
ax.set_ylabel("Sales (in $1000s)", fontsize=14, fontweight='bold', labelpad=10, color='#444')

# Rotate x-axis labels
ax.set_xticklabels(categories, rotation=30, fontsize=12, color='#222')

# Add horizontal grid lines
ax.yaxis.grid(True, linestyle='--', alpha=0.6, linewidth=0.7, zorder=0)

# Customize spines (remove top & right)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Customize left and bottom spines
ax.spines['left'].set_linewidth(1.5)
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['bottom'].set_color('black')

# Set background color
ax.set_facecolor('#f0f0f0')

# Customize tick labels
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x)}K'))  # Formatting Y-axis as "$XK"

# Add minor ticks for better readability
ax.yaxis.set_minor_locator(plt.MultipleLocator(10))
ax.tick_params(axis='y', which='minor', length=4, color='gray', width=0.8)

# Add a border to the entire chart
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)

# Add legend with a fancy border
ax.legend(["Sales Data"], loc='upper right', fontsize=12, frameon=True, edgecolor='black', shadow=True)

# Show the plot
plt.show()
```
